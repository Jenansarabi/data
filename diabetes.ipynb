```python
uploaded = files.upload()

df = pd.read_csv(io.BytesIO(uploaded['diabetes2.csv']))
print(df)
```

```python
# split the dataset into train and test sets by using balanced data
train_df = df.sample(frac=0.7, random_state=42)
test_df = df.drop(train_df.index)

# Replace the missing (zero) values of Insulin in the train set with the corresponding median value based on the Outcome value.
insulin_not_zero_outcome_1_train = train_df[(train_df['Insulin'] != 0) & (train_df['Outcome'] == 1)]
median_insulin_outcome_1_train = insulin_not_zero_outcome_1_train['Insulin'].median()
train_df.loc[(train_df['Insulin'] == 0) & (train_df['Outcome'] == 1), 'Insulin'] = median_insulin_outcome_1_train

insulin_not_zero_outcome_0_train = train_df[(train_df['Insulin'] != 0) & (train_df['Outcome'] == 0)]
median_insulin_outcome_0_train = insulin_not_zero_outcome_0_train['Insulin'].median()
train_df.loc[(train_df['Insulin'] == 0) & (train_df['Outcome'] == 0), 'Insulin'] = median_insulin_outcome_0_train
```

```python
##median with train

# Filter the dataframe to include only the rows where Insulin is not zero and Outcome is 1
insulin_not_zero_outcome_1 = train_df[(train_df['Insulin'] != 0) & (train_df['Outcome'] == 1)]

# Calculate the median of the Insulin values where Insulin is not zero and Outcome is 1
median_insulin_outcome_1 = np.median(insulin_not_zero_outcome_1['Insulin'])

# Filter the dataframe to include only the rows where Insulin is not zero and Outcome is 0
insulin_not_zero_outcome_0 = train_df[(train_df['Insulin'] != 0) & (train_df['Outcome'] == 0)]

# Calculate the median of the Insulin values where Insulin is not zero and Outcome is 0
median_insulin_outcome_0 = np.median(insulin_not_zero_outcome_0['Insulin'])

# Replace the missing (zero) values of Insulin with the corresponding median value based on the Outcome value in train set
train_df.loc[(train_df['Insulin'] == 0) & (train_df['Outcome'] == 1), 'Insulin'] = median_insulin_outcome_1
train_df.loc[(train_df['Insulin'] == 0) & (train_df['Outcome'] == 0), 'Insulin'] = median_insulin_outcome_0
# Print the median Insulin values
print('Median Insulin value for Outcome=1:', median_insulin_outcome_1)
print('Median Insulin value for Outcome=0:', median_insulin_outcome_0)

```

```python
from sklearn.model_selection import train_test_split


# Specify number of train-test splits
n_splits = 5

# Initialize lists to store median insulin values for each split
median_insulin_outcome_1_list = []
median_insulin_outcome_0_list = []

# Loop over train-test splits
for i in range(n_splits):
    # Split dataset into train and test sets
    train_set, test_set = train_test_split(df, test_size=0.2, random_state=i)
    
    # Filter the train set to include only the rows where Insulin is not zero and Outcome is 1
    insulin_not_zero_outcome_1 = train_set[(train_set['Insulin'] != 0) & (train_set['Outcome'] == 1)]

    # Calculate the median of the Insulin values where Insulin is not zero and Outcome is 1
    median_insulin_outcome_1 = np.median(insulin_not_zero_outcome_1['Insulin'])
    median_insulin_outcome_1_list.append(median_insulin_outcome_1)

    # Filter the train set to include only the rows where Insulin is not zero and Outcome is 0
    insulin_not_zero_outcome_0 = train_set[(train_set['Insulin'] != 0) & (train_set['Outcome'] == 0)]

    # Calculate the median of the Insulin values where Insulin is not zero and Outcome is 0
    median_insulin_outcome_0 = np.median(insulin_not_zero_outcome_0['Insulin'])
    median_insulin_outcome_0_list.append(median_insulin_outcome_0)

    # Replace the missing (zero) values of Insulin with the corresponding median value based on the Outcome value in train set
    train_set.loc[(train_set['Insulin'] == 0) & (train_set['Outcome'] == 1), 'Insulin'] = median_insulin_outcome_1
    train_set.loc[(train_set['Insulin'] == 0) & (train_set['Outcome'] == 0), 'Insulin'] = median_insulin_outcome_0

    # Print the median Insulin values for the current split
    print(f"Median Insulin value for Outcome=1 in split {i+1}: {median_insulin_outcome_1}")
    print(f"Median Insulin value for Outcome=0 in split {i+1}: {median_insulin_outcome_0}")
```
